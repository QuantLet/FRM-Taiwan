```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.dates as mdates
import matplotlib.path as mpath
```

```python
def create_semicircle(up=True):
    theta = np.linspace(0, np.pi, 50)  # 生成 50 個點來創建一個平滑的半圓
    x = np.cos(theta)
    y = np.sin(theta) if up else -np.sin(theta)
    verts = np.vstack([x, y]).T
    codes = [mpath.Path.MOVETO] + [mpath.Path.LINETO] * (len(verts) - 2) + [mpath.Path.CLOSEPOLY]
    return mpath.Path(verts, codes)

UpperHalfCircle = create_semicircle(up=True)
LowerHalfCircle = create_semicircle(up=False)
```

```python
start_date = pd.to_datetime('2006-01-01')
end_date = pd.to_datetime('2024-01-01')
MI = pd.read_csv('../../01 Raw Data/Others/MI.csv', index_col='Date', parse_dates=['Date'])
MI = MI.loc[(MI.index >= start_date) & (MI.index <= end_date)]
```

```python
plt.figure(figsize=(30, 6))
plt.plot(MI['MI'], color='gray', linewidth=7, alpha=0.2)
plt.axhline(y=16.5, color='black', linestyle='-.', linewidth=0.6, alpha=0.5)
plt.axhline(y=22.5, color='black', linestyle='-.', linewidth=0.6, alpha=0.5)
plt.axhline(y=31.5, color='black', linestyle='-.', linewidth=0.6, alpha=0.5)
plt.axhline(y=37.5, color='black', linestyle='-.', linewidth=0.6, alpha=0.5)
for i in range(len(MI)):
    val = MI['MI'].iloc[i]
    if val < 17:
        plt.scatter(MI.index[i], val, color='royalblue', s=80, edgecolors='black', alpha=1)
    elif val < 23:
        plt.scatter(MI.index[i], val, color='gold', s=80, edgecolors='black', alpha=1, marker=UpperHalfCircle)
        plt.scatter(MI.index[i], val, color='royalblue', s=80, edgecolors='black', alpha=1, marker=LowerHalfCircle)
    elif val < 32:
        plt.scatter(MI.index[i], val, color='forestgreen', s=80, edgecolors='black', alpha=1)
    elif val < 38:
        plt.scatter(MI.index[i], val, color='red', s=80, edgecolors='black', alpha=1, marker=UpperHalfCircle)
        plt.scatter(MI.index[i], val, color='gold', s=80, edgecolors='black', alpha=1, marker=LowerHalfCircle)
    else:
        plt.scatter(MI.index[i], val, color='red', s=80, edgecolors='black', alpha=1)
plt.gca().xaxis.set_major_locator(mdates.YearLocator(1))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(fontsize=14, fontweight='bold')
plt.yticks(fontsize=14, fontweight='bold')
plt.ylim(8, 46)
plt.tight_layout()
plt.savefig('MI time series plot.png', dpi=144, transparent=True)
plt.show()
```
